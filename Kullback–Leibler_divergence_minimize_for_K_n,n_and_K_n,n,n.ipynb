## **K n,n**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def objective(x, n):
    x0, x1, x2 = x
    return  x0 * np.log(x0 * 2*n) + x1 * np.log(x1 * 2) + x2 * np.log(x2 * 2 *n / (n-1))
def expectedvalue(x):
  x0, x1, x2 = x
  return x1 + x2*2


def constraint(x, p):
    return p - expectedvalue(x)


def optimize_objective(p):
    x0 = [0.5, 0.5, 0.5]
    bounds = [(0, 1), (0, 1), (0,1)]
    constraints = [{'type': 'eq', 'fun': lambda x: x[0] + x[1] + x[2] - 1},
                   {'type': 'ineq', 'fun': lambda x: constraint(x,p)}]
    result = minimize(objective, x0, args=(n,), bounds=bounds, constraints=constraints)
    return result.x





n = 3 #input n in K_n
D =1.1*(3*n-2)/(2*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt2xdata = p_values
plt2ydata = objective_values


n = 4 #input n in K_n
D =1.1*(3*n-2)/(2*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt4xdata = p_values
plt4ydata = objective_values



n = 2 #input n in K_n
D =1.1*(3*n-2)/(2*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt6xdata = p_values
plt6ydata = objective_values

plt_Kn = plt.plot(p_values, objective_values)
plt.xlabel('p')
plt.ylabel('Objective Function')
plt.title('Objective Function vs p')
plt.show()
```

## **K n,n,n**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def objective(x, n):
    x0, x1, x2 = x
    return  x0 * np.log(x0 * 3*n) + x1 * np.log(x1 * 1.5) + x2 * np.log(x2 * 3 *n / (n-1))

def expectedvalue(x):
  x0, x1, x2 = x
  return x1 + x2*2


def constraint(x, p):
    return p - expectedvalue(x)


def optimize_objective(p):
    x0 = [0.5, 0.5, 0.5]
    bounds = [(0, 1), (0, 1), (0,1)]
    constraints = [{'type': 'eq', 'fun': lambda x: x[0] + x[1] + x[2] - 1},
                   {'type': 'ineq', 'fun': lambda x: constraint(x,p)}]
    result = minimize(objective, x0, args=(n,), bounds=bounds, constraints=constraints)
    return result.x





n = 4 #input n in K_n
D =1.1*(4*n-2)/(3*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt1xdata = p_values
plt1ydata = objective_values

n = 8 #input n in K_n
D =1.1*(4*n-2)/(3*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt3xdata = p_values
plt3ydata = objective_values



n = 2 #input n in K_n
D =1.1*(4*n-2)/(3*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt5xdata = p_values
plt5ydata = objective_values


plt_Kn = plt.plot(p_values, objective_values)
plt.xlabel('p')
plt.ylabel('Objective Function')
plt.title('Objective Function vs p')
plt.show()
```
