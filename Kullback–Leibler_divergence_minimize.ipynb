## **Compelete Graph**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def objective(x, n):
    x1, x2 = x
    return x1 * np.log(x1 * n) + x2 * np.log(x2 * n / (n - 1))

def constraint(x, p):
    x1, x2 = x
    return p - x2

def optimize_objective(p):
    x0 = [0.5, 0.5]
    bounds = [(0, 1), (0, 1)]
    constraints = [{'type': 'eq', 'fun': lambda x: x[0] + x[1] - 1},
                   {'type': 'ineq', 'fun': lambda x: constraint(x,p)}]
    result = minimize(objective, x0, args=(n,), bounds=bounds, constraints=constraints)
    return result.x



n = 10 #input n in K_n
D =1.1*(n-1)/n

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt1xdata = p_values
plt1ydata = objective_values

plt_Kn = plt.plot(p_values, objective_values)
plt.xlabel('p')
plt.ylabel('Objective Function')
plt.title('Objective Function vs p')
plt.show()
```

## **Compelete Graph without hamiltonian cycle**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def objective(x, n):
    x1, x2, x3 = x
    return x1 * np.log(x1 * n) + x2 * np.log(x2 * n / (n - 3)) + x3 * np.log(x3*n/2)

def constraint(x, p):
    x1, x2, x3 = x
    return p - x3*2-x2

def optimize_objective(p):
    x0 = [0.5, 0.5, 0.5]
    bounds = [(0, 1), (0, 1), (0,1)]
    constraints = [{'type': 'eq', 'fun': lambda x: x[0] + x[1] + x[2] - 1},
                   {'type': 'ineq', 'fun': lambda x: constraint(x,p)}]
    result = minimize(objective, x0, args=(n,), bounds=bounds, constraints=constraints)
    return result.x



n = 10 #input n in K_n
D = 1.1 * (n+1)/n #D bar calculation

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt2xdata = p_values
plt2ydata = objective_values



plt_CnC = plt.plot(p_values, objective_values)
plt.xlabel('p')
plt.ylabel('Objective Function')
plt.title('Objective Function vs p')
plt.show()
```

## **C6**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def objective(x):
    x1, x2, x3, x4 = x
    return x1 * np.log(x1 * 6) + x2 * np.log(x2 * 6 / (2)) + x3 * np.log(x3 * 6/ 2) + x4 * np.log(x4*6)

def expectedvalue(x,diam):
  s = 0
  for i in range(0,diam + 1,1):
    s = s + i * x[i]
  return s

def constraint(x, p, diam):
    return p - expectedvalue(x,diam)

def optimize_objective(p, diam):
    x0 = [0.5, 0.5, 0.5, 0.5]
    bounds = [(0, 1), (0, 1), (0,1), (0,1)]
    constraints = [{'type': 'eq', 'fun': lambda x: x[0] + x[1] + x[2] + x[3] - 1},
                   {'type': 'ineq', 'fun': lambda x: constraint(x,p, diam)}]
    result = minimize(objective, x0, bounds=bounds, constraints=constraints)
    return result.x



n = 6 #input n in K_n

diam = 3 #input

D = 1.1*(diam*diam)/n #D bar calculation

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p, diam)) for p in p_values]

plt3xdata = p_values
plt3ydata = objective_values

plt_C6 = plt.plot(p_values, objective_values)
plt.xlabel('p')
plt.ylabel('Objective Function')
plt.title('Objective Function vs p')
plt.show()
```

## **K n,n**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def objective(x, n):
    x0, x1, x2 = x
    return  x0 * np.log(x0 * 2*n) + x1 * np.log(x1 * 2) + x2 * np.log(x2 * 2 *n / (n-1))
def expectedvalue(x):
  x0, x1, x2 = x
  return x1 + x2*2


def constraint(x, p):
    return p - expectedvalue(x)


def optimize_objective(p):
    x0 = [0.5, 0.5, 0.5]
    bounds = [(0, 1), (0, 1), (0,1)]
    constraints = [{'type': 'eq', 'fun': lambda x: x[0] + x[1] + x[2] - 1},
                   {'type': 'ineq', 'fun': lambda x: constraint(x,p)}]
    result = minimize(objective, x0, args=(n,), bounds=bounds, constraints=constraints)
    return result.x





n = 10 #input n in K_n
D =1.1*(3*n-2)/(2*n)

# Plotting the objective function
p_values = np.linspace(0, D, 100)
objective_values = [objective(optimize_objective(p), n) for p in p_values]

plt5xdata = p_values
plt5ydata = objective_values

plt_Kn = plt.plot(p_values, objective_values)
plt.xlabel('p')
plt.ylabel('Objective Function')
plt.title('Objective Function vs p')
plt.show()
```
